```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the vine_table.csv
vine_table = pd.read_csv("vine_table.csv")
```

```python
# Convert vine_table to DataFrame
vine_df = pd.DataFrame(vine_table)
vine_df
```

```python
# Filter the dataframe with totle_votes has equal or greater than 20
filtered_vine_df = vine_df[(vine_df['total_votes']) >= 20]
filtered_vine_df
```

```python
# Filter the dataframe that helpful_vote divided by total_votes is greater or equal than 50%
filtered_vine_df = filtered_vine_df[(filtered_vine_df['helpful_votes']/filtered_vine_df['total_votes'] >= 0.5)]
filtered_vine_df
```

```python
# Create a DataFrame which has reviews as part of the vine program
vine_program_df = filtered_vine_df[(filtered_vine_df['vine'] == 'Y')]
vine_program_df
```

```python
# Create a DataFrame which has reviews NOT as part of the vine program
not_vine_program_df = filtered_vine_df[(filtered_vine_df['vine'] == 'N')]
not_vine_program_df
```

```python
# Determine the total number of reviews, the number of 5-star reviews, 
# and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
percentage_vine = vine_program_df.groupby(['star_rating']).count() / vine_program_df['star_rating'].count()
percentage_vine
```

```python
vine_plot = percentage_vine.plot.pie(y='review_id', 
                                     label='star_rating', 
                                     legend=False,
                                     title='Percentage of Star Rating for Vine Program',
                                     autopct='%1.0f%%')
```

```python
percentage_not_vine = not_vine_program_df.groupby(['star_rating']).count() / not_vine_program_df['star_rating'].count()
percentage_not_vine
```

```python
not_vine_plot = percentage_not_vine.plot.pie(y='review_id', 
                                             label='star_rating', 
                                             legend=False, 
                                             title='Percentage of Star Rating for Non-Vine Program',
                                             autopct='%1.0f%%')
```
